```python
#  SPLITTING THE PROD_DESC INTO FEATURES LIKE PRODUCT TYPE, GENDER, FIT, COLOR, SIZE
```

```python
import re

def split_prod_desc(prod_desc):
    # Regex patterns for different components
    product_type_pattern = re.compile(r'\)\s(.*?)\s(?:W\'s|M\'s|Women\'s|Men\'s|Mn\'s|Ms|Mrs|M|Ms)')
    gender_pattern = re.compile(r'(W\'s|Women\'s|M\'s|Men\'s|Mn\'s)')
    fit_pattern = re.compile(r'(Reg|Petite|Pet|Tall|TrdFt|MR|R)')
    color_pattern = re.compile(r'(\w+\s\w+)$')
    size_pattern = re.compile(r'(\bXS\b|\bS\b|\bM\b|\bL\b|\bXL\b|\bXXL\b|\bXXXL\b|\d+(\.\d+)?x\d+(\.\d+)?)$')

    product_type_match = product_type_pattern.search(prod_desc)
    gender_match = gender_pattern.search(prod_desc)
    fit_match = fit_pattern.search(prod_desc)
    color_match = color_pattern.search(prod_desc)
    size_match = size_pattern.search(prod_desc)
    
    product_type = product_type_match.group(1) if product_type_match else 'Unknown'
    gender = gender_match.group(1) if gender_match else 'Unspecified'
    fit = fit_match.group(1) if fit_match else 'Other'
    color = color_match.group(1) if color_match else 'Unknown'
    size = size_match.group(1) if size_match else 'Unknown'

    # Normalize some values
    if gender == "Mn's":
        gender = "Men's"
    if fit == "TrdFt":
        fit = "Traditional Fit"
    elif fit == "Pet":
        fit = "Petite"
    elif fit == "MR":
        fit = "Men's Regular"
    elif fit == "R":
        fit = "Regular"
    
    return pd.Series([product_type, gender, fit, color, size])



# Apply the split function to PROD_DESC
df3_unique_prod_id[['Product_Type', 'Gender', 'Fit', 'Color', 'Size']] = df3_unique_prod_id['PROD_DESC'].apply(split_prod_desc)
```

```python
# ######ENCODING THE FEATURES AND COMPUTING THE MATRIX USING COSINE SIMILARITY ########
```

```python
import pandas as pd
from sklearn.preprocessing import OneHotEncoder
from sklearn.metrics.pairwise import cosine_similarity

# Encoding categorical variables
encoder = OneHotEncoder()
encoded_features = encoder.fit_transform(df3_unique_prod_id[['Product_Type', 'Gender', 'Fit', 'Color', 'Size']])

# Compute the Cosine Similarity matrix
cosine_sim = cosine_similarity(encoded_features, encoded_features)

# Mapping indices and product IDs
indices = pd.Series(df3_unique_prod_id.index, index=df3_unique_prod_id['PROD_ID']).drop_duplicates()
```

```python
# FUNCTION THAT CAN PROVIDE SUBSTITUTIONS FOR A GIVEN PRODUCT ID BASED ON THEIR SIMILARITY SCORE
```

```python
def get_recommendations_with_description(prod_id, df3_unique_prod_id, cosine_sim):
    idx = indices[prod_id]  # ensure indices is properly defined to map product IDs to DataFrame indices

    # Get the pairwise similarity scores of all products with that product
    sim_scores = list(enumerate(cosine_sim[idx]))

    # Sort the products based on the similarity scores
    sim_scores = sorted(sim_scores, key=lambda x: x[1], reverse=True)

    sim_scores = sim_scores[:5]  # Include the first entry for comparison

    # Get the product indices
    product_indices = [i[0] for i in sim_scores]

    # Get the similarity scores (for visibility)
    similarity_scores = [i[1] for i in sim_scores]

    # Return the top 10 most similar products and their descriptions from the correct DataFrame
    result = df3_unique_prod_id.loc[product_indices, ['PROD_ID', 'PROD_DESC']]
    result['Similarity'] = similarity_scores
    return result
```

```python
###### ENCODING THE FEATURES AND COMPUTING THE MATRIX USING EUCLIDEAN DISTANCE ########
```

```python
from sklearn.metrics.pairwise import euclidean_distances


encoder = OneHotEncoder()
encoded_features = encoder.fit_transform(df3_unique_prod_id[['Product_Type', 'Gender', 'Fit', 'Color', 'Size']])

# Compute the Euclidean Distance matrix
euclidean_dist = euclidean_distances(encoded_features, encoded_features)

# Mapping indices and product IDs
indices = pd.Series(df3_unique_prod_id.index, index=df3_unique_prod_id['PROD_ID']).drop_duplicates()
```

```python
def get_recommendations_with_description(prod_id, df3_unique_prod_id, euclidean_dist):
    idx = indices[prod_id]  # ensure indices is properly defined to map product IDs to DataFrame indices

    # Get the pairwise distance scores of all products with that product
    dist_scores = list(enumerate(euclidean_dist[idx]))

    # Sort the products based on the distance scores (ascending order)
    dist_scores = sorted(dist_scores, key=lambda x: x[1])

    dist_scores = dist_scores[:5]  # Include the first entry for comparison

    # Get the product indices
    product_indices = [i[0] for i in dist_scores]

    # Get the distance scores (for visibility)
    distance_scores = [i[1] for i in dist_scores]

    # Return the top 5 most similar products and their descriptions from the correct DataFrame
    result = df3_unique_prod_id.loc[product_indices, ['PROD_ID', 'PROD_DESC']]
    result['Distance'] = distance_scores
    return result
```
